```python
#Import Dependencies
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

#Set up file path
cancer_path=Path('Resources/cancer_dataset.csv')

#Read in the csv
cancer_df=pd.read_csv(cancer_path,encoding="ISO-8859-1")

#Converting three columns to integars for calculations.
cancer_df=cancer_df.astype({"Percent_of_Population_Married":int,"Average_Deaths_per_Year":int,"Population_Estimate_2015":int,"Median_Age":int})
```

```python
#Creating new column called Percent Cancer Deaths.
cancer_df["Percent_Cancer_Deaths"]=(cancer_df["Average_Deaths_per_Year"]/cancer_df["Cancer_Incidence_Count"])*100
cancer_df.head(20)
```

```python
#Seeting up bins for categories of ages and making a copy of the original dataset.
age_bins=[20,25,30,35,40,45,50,55,60,65]
labels=["20-25","25-30","30-35","35-40","40-45","45-50","50-55","55-60","60-65"]
age_df=cancer_df.copy()
```

```python
#Binning the data by age.
age_df["Median_Age"]=pd.cut(age_df["Cancer_Incidence_Count"],age_bins,labels=labels)

#Finding the average marriage percent in each category of ages.
age_marriage=age_df.groupby(["Median_Age"])["Percent_of_Population_Married"].mean()
```

```python
#Adding the marriage percents to a dataframe.
age_and_marriage=pd.DataFrame({"Marriage Percent":age_marriage})
age_and_marriage
```

```python
#Plotting a barchart showing the average marriage percent for each category of ages.
age_and_marriage.plot(kind="bar",figsize=(10,6),color='r')
#Chart labels
plt.title("Marriage Percent by Age") 
plt.xlabel("Median Age")
plt.ylabel("Marriage Percent")
plt.legend()
```
